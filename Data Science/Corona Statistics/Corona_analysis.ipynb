## **Analysis of the Data-Frame**

```python
df.describe()
```

 Highest population of the country

```python
countries=df["Country/Region"]
population=df["Population"]
max_index = population.idxmax()
largest_country = countries[max_index]
largest_population = population[max_index]

print(f"The country with the largest population is {largest_country} with a population of {largest_population}.")
```

Highest test performed



```python
Total_tests=df["TotalTests"]
highest_test_index=Total_tests.idxmax()
country_with_highest_test= countries[highest_test_index]
highest_test_number = Total_tests[highest_test_index]

print(f"The country with the Highest test is {country_with_highest_test} with a count of {highest_test_number}.")
```

Highest deadth in a country

```python
deadth_column=df["TotalDeaths"]
maximum_deadth_index=deadth_column.idxmax()
highest_deadth_country=countries[maximum_deadth_index]
highest_deadth_number=deadth_column[maximum_deadth_index]
print(f"The country with the Highest deadth is {highest_deadth_country} with a count of {highest_deadth_number}.")
```

 **Finding the test per Million**

```python
df['TestsPer-Million'] = (df['TotalTests'] / df['Population']) * 1_000_000
```

```python
test_per_million_column=df["TestsPer-Million"]
highest_test_PM_index=test_per_million_column.idxmax()
lowest_test_PM_index=test_per_million_column.idxmin()
country_highest_testPM=countries[highest_test_PM_index]
country_lowest_testPM=countries[lowest_test_PM_index]
print(f'country with the highest test according to the population is {country_highest_testPM} with count {test_per_million_column[highest_test_PM_index]}')
print(f'country with the lowest test according to the population is {country_lowest_testPM} with count {test_per_million_column[lowest_test_PM_index]}')
```

# **Probabilistic modeling and Reasoning**

## **Survival Rate according the the Total cases observed**

```python
# Calculate the survival rate for each country
df['SurvivalRate'] = (1 - (df['TotalDeaths'] / df['TotalCases'])) * 100

# Sort by the lowest survival rate (which implies the highest mortality rate)
lowest_survival_rates = df[['Country/Region', 'SurvivalRate']].sort_values(by='SurvivalRate').head(10)

# Display the countries with the lowest survival rates
lowest_survival_rates
```

## **Taking the probability on the  basis of total population**

```python
# Calculate the survival rate for each country
df['SurvivalRate'] = (1 - (df['TotalDeaths'] / df['Population'])) * 100

# Sort by the lowest survival rate (which implies the highest mortality rate)
lowest_survival_rates = df[['Country/Region', 'SurvivalRate']].sort_values(by='SurvivalRate').head(10)

# Display the countries with the lowest survival rates
lowest_survival_rates
```

## **Visualyzing the dataset**

```python

# Set up the plotting style

sns.set(style="whitegrid")
```

```python
# 1. Bar chart for the top 10 countries by total deaths
top_10_deaths = df[['Country/Region', 'TotalDeaths']].sort_values(by='TotalDeaths', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x='TotalDeaths', y='Country/Region', data=top_10_deaths, palette='Reds_r')
plt.title('Top 10 Countries by Total Deaths')
plt.xlabel('Total Deaths')
plt.ylabel('Country/Region')
plt.show()
```

```python
# 2. Bar chart for the top 10 countries by total tests
top_10_tests = df[['Country/Region', 'TotalTests']].sort_values(by='TotalTests', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x='TotalTests', y='Country/Region', data=top_10_tests, palette='Blues_r')
plt.title('Top 10 Countries by Total Tests')
plt.xlabel('Total Tests')
plt.ylabel('Country/Region')
plt.show()
```

```python

# 3. Bar chart comparing tests per million population across a selection of countries
selected_countries = df[['Country/Region', 'Tests/1M pop']].sort_values(by='Tests/1M pop', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x='Tests/1M pop', y='Country/Region', data=selected_countries, palette='Greens_r')
plt.title('Top 10 Countries by Tests per Million Population')
plt.xlabel('Tests per Million Population')
plt.ylabel('Country/Region')
plt.show()
```

```python
# 4. Scatter plot to compare total cases and total deaths
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TotalCases', y='TotalDeaths', hue='Continent', data=df, palette='Set2', s=100)
plt.title('Total Cases vs Total Deaths')
plt.xlabel('Total Cases')
plt.ylabel('Total Deaths')
plt.show()
```

```python


# 5. Pie chart showing the distribution of cases by continent
continent_cases = df.groupby('Continent')['TotalCases'].sum().reset_index()
plt.figure(figsize=(8, 8))
plt.pie(continent_cases['TotalCases'], labels=continent_cases['Continent'], autopct='%1.1f%%', colors=sns.color_palette('Set2'))
plt.title('Distribution of Cases by Continent')
plt.show()
```

Total Cases Vs Total Death

```python
# Select the top 10 countries by total cases
top_10_cases = df[['Country/Region', 'TotalCases', 'TotalDeaths']].sort_values(by='TotalCases', ascending=False).head(10)

# Scatter plot to compare total cases and total deaths for the top 10 countries
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TotalCases', y='TotalDeaths', data=top_10_cases, hue='Country/Region', palette='Set2', s=100)
plt.title('Total Cases vs Total Deaths (Top 10 Countries)')
plt.xlabel('Total Cases')
plt.ylabel('Total Deaths')
plt.legend(title='Country/Region')
plt.show()
```

## Mortality Rate

```python
# Calculate mortality rate for each country
df['MortalityRate'] = (df['TotalDeaths'] / df['TotalCases']) * 100

# Sort countries by mortality rate
top_10_mortality = df[['Country/Region', 'MortalityRate']].sort_values(by='MortalityRate', ascending=False).head(10)

# Plotting the top 10 countries by mortality rate
plt.figure(figsize=(12, 6))
sns.barplot(x='MortalityRate', y='Country/Region', data=top_10_mortality, palette='Oranges_r')
plt.title('Top 10 Countries by Mortality Rate (%)')
plt.xlabel('Mortality Rate (%)')
plt.ylabel('Country/Region')
plt.show()
```
