```python
import matplotlib.pyplot as plt

x=["Python","C","C++","Java","Javascript"]
y=[89,84,82,88,80]

plt.bar(x,y,edgecolor='r',color='y')
plt.title("Bar Graph")
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.show()
```
