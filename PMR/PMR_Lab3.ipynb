# **Titanic EDA**

### **Step 1: Loading the Dataset**

```python
import seaborn as sns
import pandas as pd

titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

### **Step 2: Data Overview**

```python
#Basic Information about the dataset
titanic.info()

titanic.describe(include='all')
```

```python
titanic.describe()
```

### **Step 3: Handling the missing values**

```python
missing_values=titanic.isnull().sum()
missing_values[missing_values>0]


#filling missing age value with the median
titanic['age'].fillna(titanic['age'].median(),inplace=True)

#filling missing 'embarked' value with the mode
titanic['embarked'].fillna(titanic['embarked'].mode()[0],inplace=True)

#Dropping the rows where 'deck' have the missing values
titanic.drop(columns=['deck'],inplace=True)

#dropping the rowz where the 'embark_town' has the missing values,as it is a categorial feature
titanic.dropna(subset=['embark_town'],inplace=True)
```

### **Data VIsulaization**
