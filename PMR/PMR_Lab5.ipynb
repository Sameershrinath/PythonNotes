## **Box Plot**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 


#General random data for  the box plot 
np.random.seed(42)
data=[np.random.normal(50,10,100),np.random.normal(70,15,100),np.random.normal(60,5,100)]

#create the box plot 
plt.figure(figsize=(8,6))
sns.boxplot(data=data)
plt.xticks([0,1,2],['Group 1','Group 2','Group 3'])
plt.title('Box Plot')
plt.ylabel('values')
plt.show()
```

## **Dot Plot [strip Plot]**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
import pandas as pd 


# Sample data for the dot plot 
np.random.seed(42)
data={
    'category': ['A','B','C','D','E'],
    'Values':np.random.randint(10,100,size=5)
    }
df=pd.DataFrame(data)

#create a dot plot (strip plot)
plt.figure(figsize=(8,6))
sns.stripplot(x='category',y='Values',data=df,size=10,jitter=True,palette='Set2')
plt.title('Dot Plot')
plt.ylabel('Values')
plt.show()

```

## **Bar Chart**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
import numpy as np 


# sample data for the bar chart
np.random.seed(42)
data={
    'category': ['A','B','C','D','E'],
    'Values':np.random.randint(10,100,size=5)
    }
df=pd.DataFrame(data)


# create the bar chart 
plt.figure(figsize=(8,6))
sns.barplot(x='category',y='Values',data=df,palette='Blues_d')
plt.title("Bar Chart")
plt.ylabel('Values')
plt.show()
```
