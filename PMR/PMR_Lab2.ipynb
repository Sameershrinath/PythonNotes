# **Example 1: Basic Probability Calculation**

```python
import random

#probability of  an event
def probability(event_outcome,total_outcome):
    return event_outcome/total_outcome

#probability of rolling a 3 on 6 sided dice
event_outcome=1
total_outcome=6 
prob=probability(event_outcome,total_outcome)
print(f"Probability of rolling 3 is {prob}")
```

## **Example 2: Simulating a coin Toss**

```python
def coin_toss():
    return random.choice(['Heads','Tails'])
tosses=[coin_toss() for _ in range(1000)]
head_count=tosses.count('Heads')
tail_count=tosses.count('Tails')
print(f'Heads: {head_count},Tails:{tail_count}')
```

```python
import matplotlib.pyplot as plt
list2=[head_count,tail_count]
list3=["head","Tail"]
plt.bar(list3,list2)
plt.xlabel("Outcome")
plt.ylabel("NO. of outcome")
plt.title("simulation of 1000 rolls")
plt.show()
```

## **Example 3: Probability Distribution (Discrete)**

```python
import matplotlib.pyplot as plt 
# probability distribution of the 6 sided dice
outcomes=[1,2,3,4,5,6]
probability=[1/6]*6 #Uniform distribution 
plt.bar(outcomes,probability)
plt.xlabel('Die Outcome')
plt.ylabel('probability')
plt.title('Probability distrubbution of a Six-Sided Die')
plt.show()
```

## **Example 4: Simulating Random Events (Monte Carlo Simulation)**

```python
import numpy as np
def roll_dice():
    return random.randint(1,6),random.randint(1,6)
def simulate_dice_rolls(num_simulations):
    success_count=0
    for _ in range(num_simulations):
        roll1,roll2=roll_dice()
        if roll1 + roll2 == 7:
            success_count+=1
    return success_count/num_simulations

#simulate the 10000 rolls
num_simulations=10000
estimated_prob=simulate_dice_rolls(num_simulations)
print(f'Estimated Probability of rolling a sum of 7: {estimated_prob}')
```

```python
#tossing a coin 1000 times 
import random

def tossing_coin():
    return random.choice(["head","tail"])

result=[tossing_coin() for _ in range(1000)]
heads=result.count("head")
tail=result.count('tail')

print(f'The total no. of head is {heads} and tail is {tail}')
```

```python
#probability of getting head 

# no of favourable outcomes/ no of total outcomes

def probability(fav,total):
    return fav/total

Probability_of_getting_head=probability(heads,1000)
Probability_of_getting_tail=probability(tail,1000)

print(f'The probability of getting head is {Probability_of_getting_head}')
print(f'The probability of getting tail is {Probability_of_getting_tail}')
```
